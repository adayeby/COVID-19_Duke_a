```python
#Rewrite tencent mobility data file with city+code in BaiduMobilityDirectory

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import csv
import os
```

```python
cityreader = pd.read_csv("BaiduMobilityDirectory.csv")
print(cityreader)
```

```python
tencentreader = pd.read_csv("data/Mobility_chunyun2018.csv")
print(len(tencentreader))
print(tencentreader)
```

```python
def modifyExcep(city):
    newCity = ""
    if "'" in city:
        idx = city.find("'")
        city = city[:idx]+city[idx+1:]
    if city == "Jinzhou (Liaoning)":
        newCity = "Jinzhou_210700"
        return newCity
    elif city == "Yichun (Heilongjiang)":
        newCity = "Yichun_230700"
        return newCity
    elif city == "Suzhou (Jiangsu)":
        newCity = "Suzhou_320500"
        return newCity
    elif city == "Taizhou (Jiangsu)":
        newCity = "Taizhou_321200"
        return newCity
    elif city == "Taizhou (Zhejiang)":
        newCity = "Taizhou_331000"
        return newCity
    elif city == "Suzhou (Anhui)":
        newCity = "Suzhou_341300"
        return newCity
    elif city == "Fuzhou (Fujian)":
        newCity = "Fuzhou_350100"
        return newCity
    elif city == "Yichun (Jiangxi)":
        newCity = "Yichun_360900"
        return newCity
    elif city == "Fuzhou (Jiangxi)":
        newCity = "Fuzhou_361000"
        return newCity
    #No Jinzhou (Hubei)
    elif city == "Liuzhou (Guangxi)":
        newCity = "Liuzhou_450200"
        return newCity
    elif city == "Wuzhou (Guilin)":
        newCity = "Wuzhou_450400"
        return newCity
    elif city == "Yulin (Guangxi)":
        newCity = "Yulin_450900"
        return newCity
    elif city == "Suining (Sichuan)":
        newCity = "Suining_510900"
        return newCity
    #No Bazhou
    elif city == "Yulin (Shanxi)":
        newCity = "Yulin_610800"
        return newCity
    #No Wuzhou (Ningxia)
    #No Bozhou (Xinjiang)
    elif city == "Hetian (Xinjiang)" or city == "Hetianxian (Xinjiang)":
        newCity = "Hetiandiqu_653200"
        return newCity
    #No Wujiaqu (Xinjiang) or Wujiaqushi (Xinjiang)
    elif city == "Nannig":
        newCity = "Nanning_450100"
        return newCity
    elif city == "Xintai":
        newCity = "Xingtai_130500"
        return newCity
    elif city == "Saming":
        newCity = "Sanming_350400"
        return newCity
    elif city == "Urumuqi":
        newCity = "Wulumuqi_650100"
        return newCity
    elif city == "Kashi":
        newCity = "Kashendiqu_653100"
        return newCity
    elif city == "zhanjiang":
        newCity = "Zhanjiang_440800"
        return newCity
    else:
        for c in cityreader["City_EN"]:
            if city == c[:-7]:
                newCity = c
                return newCity
        for c in cityreader["City_EN"]:
            if city in c:
                newCity = c
                return newCity
    return city
```

```python
cityDict = {}
newCity = ""
fullSet = set(tencentreader["Origin"])|set(tencentreader["Destination"])
for city in fullSet:
    newCity = modifyExcep(city)
    cityDict[city] = newCity 
    
print(cityDict["Xi'an"])
```

```python
filename = "Mobility_chunyun2018_citycode.csv"
fields = ["Day","Origin","Destination","Mobility Index"]
with open(filename, 'w') as csvfile: 
    csvwriter = csv.writer(csvfile) 
    csvwriter.writerow(fields) 
    for i in range(len(tencentreader)):
        day = tencentreader.loc[i,"Day"]
        ori = cityDict[str(tencentreader.loc[i,"Origin"])]
        dest = cityDict[str(tencentreader.loc[i,"Destination"])]
        mi = tencentreader.loc[i,"Mobility Index"]
        csvwriter.writerow([day,ori,dest,mi])
        if i == 92247:
            print("complete!")
    
```

```python
chunyunCitycode = pd.read_csv("Mobility_chunyun2018_citycode.csv")
print(chunyunCitycode)
```

```python
import datetime as dt

date = []
dayJan = 10
daydict = {}
for num in range(1,15):
    d = dt.datetime(2020,1,dayJan)
    daydict[num] = d.strftime("%x")
    dayJan += 1

for i in chunyunCitycode.loc[:,"Day"]:
    date.append(daydict[i])

chunyunCitycode["date"] = date
print(chunyunCitycode)
```
