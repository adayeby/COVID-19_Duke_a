```python
#create new Baidu Mobility Out data
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
import csv
import os
import datetime as dt
```

```python
bdreader = pd.read_csv("baidu_chunyun.csv")
print(bdreader.loc[0,:])
```

```python
totalreader = pd.read_csv("predicted_total_outflow.csv")
print(totalreader)
```

```python
citylst = set(totalreader["origin"])
def processBD(r):
    row = bdreader.loc[r,:]
    if row["city_out"] in citylst:
        total = totalreader[(totalreader["date"] == row["date"]) & (totalreader["origin"] == row["city_out"])]["total_est"]
        return float((row["X."]/100)*total)
    return None

print(processBD(2983))
```

```python
predicted_flow = []
for i in range(len(bdreader)):
    predicted_flow.append(processBD(i))
    if (i == 1):
        print(predicted_flow)
    if (i == len(bdreader)//2):
        print("half done!")
    if (i == len(bdreader)-1):
        print("complete!")
```

```python
bdreader["predicted_flow"] = predicted_flow
```
